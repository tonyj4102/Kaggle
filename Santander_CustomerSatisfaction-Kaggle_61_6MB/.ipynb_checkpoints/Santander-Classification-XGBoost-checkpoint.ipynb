```r
##Read in data
train=read.csv("train.csv")
test=read.csv("santander_test.csv")
```

```r
##### Removing IDs
train$ID <- NULL
test.id <-test$ID
test$ID <- NULL
```

```r
##Cleaning
```

```r
##### Removing constant features
cat("\n## Removing the constants features.\n")
for (f in names(train)) {
  if (length(unique(train[[f]])) == 1) {
    cat(f, "is constant in train. We delete it.\n")
    train[[f]] <- NULL
    test[[f]] <- NULL
  }
}
```

```r
##### Removing identical features
features_pair <- combn(names(train), 2, simplify = F)
toRemove <- c()
for(pair in features_pair) {
  f1 <- pair[1]
  f2 <- pair[2]
  
  if (!(f1 %in% toRemove) & !(f2 %in% toRemove)) {
    if (all(train[[f1]] == train[[f2]])) {
      cat(f1, "and", f2, "are equals.\n")
      toRemove <- c(toRemove, f2)
    }
  }
}

```

```r
# More cleaning -K-means
library(stats)
```

```r
# Determine the number of clusters
#create set of number of clusters from 1 to 15
wss<-numeric(8)

#find wss (within-cluster sum of squares) for each kmeans operation for i clusters
for (i in 1:8) wss [i] <-sum(kmeans(train, centers=i)$withinss)
```

```r
#plot 
plot(1:8, wss, type="b", xlab="No. of clusters", ylab="Within groups sum of squares")
```

```r
# Add new columnn by cluster number for train and test

km<-kmeans(test,3,100)
test$cluster <-km$cluster
```

```r
#sample 10% 
set.seed(113)  #15071
spl = sample(nrow(train), 1*nrow(train))
train = train[spl,]
```

```r
# split data set into train & test
set.seed(113)  #15071
spl = sample(nrow(train), 0.8*nrow(train))
Strain = train[spl,]
Stest = train[-spl,]
```

```r
#Load X-GBoost modules
require(xgboost)
```

```r
# create Y(response variable) and remove from train
y=Strain$TARGET
```

```r
Strain$TARGET<-NULL
```

```r
summary(y)
```

```r
trainMatrix <- as.matrix(Strain)
testMatrix <- as.matrix(Stest)
```

```r
#Cross Validation to find the optimum number of rounds

watchlist <- list(train=trainMatrix)

numberOfClasses <- max(y) + 1

param <- list(objective = "binary:logistic",
              booster = "gbtree",
              eval_metric = "auc", #"mlogloss"
              eta = 0.01,
              max.depth = 6,
              subsample           = 0.7,
              colsample_bytree    = 0.7,
              verbose             = 1,
              watchlist           = watchlist,
              "num_class" = numberOfClasses)

cv.nround <- 100
cv.nfold <- 3

bst.cv = xgb.cv(param=param, data = as.matrix(trainMatrix), label = y, 
                nfold = cv.nfold, nrounds = cv.nround, early.stop.round=NULL)
```

```r
# plot (AUC)
plot((bst.cv$test.auc.mean),type="l")
```

```r
# plot the AUC for the training and testing samples
library(ggplot2)
library(reshape)

dst<-bst.cv
dst<dst$ID

for (i in 1:nrow(dst)) {
    dst$ID [i]<-i
}
summary(dst)
str(dst)
auc.long <- melt(dst, id = "ID", measure = c("test.auc.mean", "train.auc.mean"))
ggplot(data=auc.long, aes(ID,value,color = variable)) + 
  geom_line() + 
  theme_bw()
```

```r
bst <- xgboost(data = trainMatrix, label = y, booster = "gbtree", max.depth = 6, eta=0.01, 
               nrounds = 100,
               nthread = 2, objective = "binary:logistic",gamma=1, eval_metric = "auc")
```

```r
#predict
preds=predict(bst,testMatrix)
```
