```r
##Cleaning
```

```r
##### Removing constant features
cat("\n## Removing the constants features.\n")
for (f in names(train)) {
  if (length(unique(train[[f]])) == 1) {
    cat(f, "is constant in train. We delete it.\n")
    train[[f]] <- NULL
    test[[f]] <- NULL
  }
}
```

```r
##### Removing identical features
features_pair <- combn(names(train), 2, simplify = F)
toRemove <- c()
for(pair in features_pair) {
  f1 <- pair[1]
  f2 <- pair[2]
  
  if (!(f1 %in% toRemove) & !(f2 %in% toRemove)) {
    if (all(train[[f1]] == train[[f2]])) {
      cat(f1, "and", f2, "are equals.\n")
      toRemove <- c(toRemove, f2)
    }
  }
}

```

```r
#remove features

feature.names <- setdiff(names(train), toRemove)

train <- train[, feature.names]
#test <- test[, feature.names]
```

```r
#sample 10% 
set.seed(113)  #15071
spl = sample(nrow(train), 0.1*nrow(train))
trainS = train[spl,]
```

```r
# split data set into train & test
set.seed(113)  #15071
spl = sample(nrow(trainS), 0.8*nrow(trainS))
Strain = trainS[spl,]
Stest = trainS[-spl,]
```

```r
#Load X-GBoost modules
require(xgboost)
```

```r
# create Y(response variable) and remove from train
y=Strain$TARGET
```

```r
Strain$TARGET<-NULL
```

```r
head(y)
```

```r
summary(y)
```

```r
trainMatrix <- as.matrix(Strain)
testMatrix <- as.matrix(Stest)
```

```r
# Cross Validation to find the optimum number of rounds
bst.cv = xgb.cv(data = as.matrix(trainMatrix), 
                label = y, 
                booster = "gbtree", 
                max.depth = 6,#7, 
                eta=0.01,# 0.125, 
                nrounds = 1000,
                gamma =1, #
                colsample_bytree = 1, #
                min_child_weight = 1, # 
                nthread = 2, 
                objective = "binary:logistic",
                eval_metric = "auc",
                nfold = 10#,
                #early.stop.round=50
)
```

```r
# plot (AUC)
plot((bst.cv$test.auc.mean),type="l")
```

```r
# plot the AUC for the training and testing samples
library(ggplot2)
library(reshape)

dst<-bst.cv
dst<dst$ID

for (i in 1:nrow(dst)) {
    dst$ID [i]<-i
}
summary(dst)
str(dst)
auc.long <- melt(dst, id = "ID", measure = c("test.auc.mean", "train.auc.mean"))
ggplot(data=auc.long, aes(ID,value,color = variable)) + 
  geom_line() + 
  theme_bw()
```

```r
bst <- xgboost(data = trainMatrix, label = y, booster = "gbtree", max.depth = 6, eta=0.01, 
               nrounds = 321,
               nthread = 2, objective = "binary:logistic",gamma=1, eval_metric = "auc")
```

```r
#predict
preds=predict(bst,testMatrix)
```

```r
str(preds)
preds[1:10]
```

```r
#AUC
#install.packages("AUC")
library(cvAUC)
auc <- AUC(preds, Stest$TARGET)
auc
```
